```python
import numpy as np
import pickle
```

```python
episode1 = pickle.load(file=open('./episode_info.p',mode='rb'))
```

```python
num_steps = len(episode1['hist_states'])
a1 = episode1['hist_action']
s1 = episode1['hist_states']

step = 0
print('Step %d: \t Episode1' % step)
print('Dist_x: \t %f' % (s1[step][0]))
print('Dist_y: \t %f' % (s1[step][1]))
print('Theta: \t\t %f' % (s1[step][2]))
print('Fill_Level: \t %f' % (s1[step][6]))
print('Spillage: \t %f' % (s1[step][7]))

for i in range(1, num_steps):
    print('---------------------------------------------')
    print('Action_x: \t %f' % (a1[i-1][0]))
    print('Action_y: \t %f' % (a1[i-1][1]))
    print('Action_theta: \t %f' % (a1[i-1][2]))
    print('---------------------------------------------')
    print('---------------------------------------------')
    print('Step %d: \t Episode1' %i)
    print('Dist_x: \t %f' % (s1[i][0]))
    print('Dist_y: \t %f' % (s1[i][1]))
    print('Theta: \t\t %f' % (s1[i][2]))
    print('Fill_Level: \t %f' % (s1[i][6]))
    print('Spillage: \t %f' % (s1[i][7]))

```
