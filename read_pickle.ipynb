```python
num_steps = len(episode2['hist_states'])
a1 = episode1['hist_action']
a2 = episode2['hist_action']
s1 = episode1['hist_states']
s2 = episode2['hist_states']

step = 0
print('Step %d: \t Episode1 \t Episode 2 ' % step)
print('Dist_x: \t %f \t %f' % (s1[step][0], s2[step][0]))
print('Dist_y: \t %f \t %f' % (s1[step][1], s2[step][1]))
print('Theta: \t\t %f \t %f' % (s1[step][2], s2[step][2]))
print('Fill_Level: \t %f \t %f' % (s1[step][6], s2[step][6]))
print('Spillage: \t %f \t %f' % (s1[step][7], s2[step][7]))

for i in range(1, num_steps):
    print('---------------------------------------------')
    print('Action_x: \t %f \t %f' % (a1[i-1][0], a2[i-1][0]))
    print('Action_y: \t %f \t %f' % (a1[i-1][1], a2[i-1][1]))
    print('Action_theta: \t %f \t %f' % (a1[i-1][2], a2[i-1][2]))
    print('---------------------------------------------')
    print('---------------------------------------------')
    print('Step %d: \t Episode1 \t Episode 2 ' %i)
    print('Dist_x: \t %f \t %f' % (s1[i][0], s2[i][0]))
    print('Dist_y: \t %f \t %f' % (s1[i][1], s2[i][1]))
    print('Theta: \t\t %f \t %f' % (s1[i][2], s2[i][2]))
    print('Fill_Level: \t %f \t %f' % (s1[i][6], s2[i][6]))
    print('Spillage: \t %f \t %f' % (s1[i][7], s2[i][7]))

```
